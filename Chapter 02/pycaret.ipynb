```python
et = create_model('et')
```

```python
tuned_et = tune_model(et)
```

```python
models()
```

```python
models(type='ensemble').index.tolist()
```

```python
bagged_et = ensemble_model(et)
```

```python
boosted_et = ensemble_model(et, method = 'Boosting')
```

```python
blender = blend_models(estimator_list = [boosted_et, bagged_et, tuned_et], method = 'soft')
```

```python
stacker = stack_models(estimator_list = [boosted_et,bagged_et,tuned_et], meta_model=et)
```
